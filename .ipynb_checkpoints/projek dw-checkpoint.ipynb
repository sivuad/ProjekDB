```python
#Koneksi databse
db = mysql.connector.connect(
    host="localhost",
    user="root",
    passwd=""
)
if db.is_connected():
    print("hasil terhubung ke Database")
    
```

```python
cursor = db.cursor()
cursor.execute("CREATE DATABASE lazada")
print("Database berhasil dibuat")
```

```python
cursor.execute("use lazada")
cursor.execute("DROP TABLE if exists kategori")
cursor.execute("DROP TABLE if exists merek")
cursor.execute("DROP TABLE if exists barang")
cursor.execute("create table kategori(id_kategori int primary key, kategori varchar(100) not null)")
cursor.execute("create table merek(id_merek int primary key, nama_merek varchar(100) not null)")
cursor.execute("create table barang(id_barang int primary key, nama_brng varchar(100) not null)")
```
